```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data:
np.random.seed(15)
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

# plot
fig, ax = plt.subplots()
VP = ax.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
import matplotlib.pyplot as plt

data = [16, 25, 47, 56, 23, 45, 19, 55, 44, 27]
stems = [1, 1, 2, 2, 2, 4, 4, 4, 5, 5]

# Criando o stem plot
plt.figure(figsize=(8, 5))  # Define o tamanho da figura
markerline, stemlines, baseline = plt.stem(stems, data, linefmt='-', markerfmt='o', basefmt=' ')

# Personalizando o gráfico
plt.ylabel('Valores (Data)', fontsize=12)
plt.xlabel('Stem', fontsize=12)
plt.title('Gráfico de Haste (Stem Plot)', fontsize=14)
plt.xlim(0, 6)  # Ajustei os limites para melhor visualização
plt.ylim(0, 60)  # Adicionei limite para o eixo y
plt.grid(True, linestyle='--', alpha=0.6)  # Adiciona grid

# Destacando os marcadores
markerline.set_markerfacecolor('red')  # Cor dos marcadores
markerline.set_markersize(8)  # Tamanho dos marcadores

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Exemplo com o dataset 'titanic' (disponível no seaborn)
dataset = sns.load_dataset('titanic')

# Gráfico swarmplot atualizado
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived', palette={0: 'red', 1: 'green'})

plt.title('Distribuição de Idade por Sexo e Sobrevivência (Titanic)')
plt.show()
```
