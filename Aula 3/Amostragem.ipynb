# Amostragem usando Pyhton

Uma amostra é um subconjunto da população em estudo. Ela representa a população maior e é usada para fazer inferências sobre essa população, ou seja, reunir informações sobre uma população, sem ter que medi-la em sua totalidade.

Existem vários tipos e formas de selecionar a amostra de uma população, desde técnicas simples até as mais complexas.

Vamos abordar tês tipos de técnicas de amostragem:
* Simple random sample – Amostragem Aleatória Simples
* Stratified sample – Amostragem Aleatória Estratificada
* Systematic sample – Amostragem Sistemática

![Amostragem tipos](https://minerandodados.com.br/wp-content/uploads/2020/03/image-3.png "Figura Amostragem")

## Amostragem Aleatória Simples

Um determinado número de elementos é retirado da população de forma aleatória, onde todos os elementos têm a mesma probabilidade de serem selecionados para a amostra.

Exemplo:

## Amostragem Aleatória Estratificada

Essa amostragem consiste em dividir toda a população em diferentes grupos ou estratos diferentes, de maneira que um indivíduo só pode fazer parte de um único grupo e nenhum desses indivíduos pode ficar de fora de um estrato, a seleção de cada um dos indivíduos é feita de forma aleatória.

Exemplo:

```python
import pandas as pd
import seaborn as sns

# Carregando a base de dados:
iris = sns.load_dataset('iris')
#outra forma:
#iris=pd.read_csv('iris.csv', sep='\t')

# Visualizando as linhas iniciais do dataframe:
iris.head()

#para salvar como csv
#iris.to_csv('iris.csv', sep='\t')
```

```python
# Informações do dataframe:
iris.info()
```

```python
# Selecionando o  número de registros de uma amostra:
df_sample = iris.sample(n=50)

# Informações do dataframe criado:
df_sample.info()
```

```python
df_sample = iris.sample(frac=0.10)

# Informações do dataframe criado:
df_sample.info()
```

## Amostragem Sistemática

Os elementos da população são colocados em uma lista e cada enésimo elemento da lista é escolhido (sistematicamente) por inclusão na amostra.

Exemplo: O diretor de uma escola usa uma lista com os nomes dos alunos e cada 10º estudante será selecionado para inclusão na amostra. O primeiro elemento deve ser selecionado aleatoriamente.

Gerando a semente aleatória

```python
import numpy as np

# Selecionado um registro aleatório entre os valores de 0 a 10:
semente = np.random.choice(10, 1)
semente
```

```python
print(semente[0])
```

Gerando índices a partir da semente

```python
# Gerando um array que inicia em 0 e termina em 150 com um intervalo de 7:
indices = np.arange(0,150,semente[0])
indices
```

Veja que o nosso array foi gerado conforme especificamos acima. Os elementos foram colocados em uma lista, o enésimo elemento escolhido foi o 7, e esses serão os valores que serão incluídos na amostra.

Observação : A semente é gerada de forma aleatória, então quando você estiver fazendo os seus experimentos o valor que será gerado pode ser diferente.

Gerando a amostra a partir dos índices

```python
amostra = iris.loc[indices,:]
amostra
```

Observe que os registros selecionados foram os que correspondem ao array de indices que criamos anteriormente.
A amostra possui 15 registros que foram gerados de forma aleatória usando o número de indices que foi criado a partir de uma semente.
